```python
import sqlite3
import json

DB_PATH = '/Users/ifditahasanorney/Documents/GitHub/multilevel_feedback_annotations/expert_annotation_interface/app-1026-usc8.db'

users = [
    [1, "6159c93ce607a8ce617fcbb7@email.prolific.com"],
    [2, "610a90daec40f207960c61ed@email.prolific.com"],
    [3, "5c18daaae467040001e68fc4@email.prolific.com"],
    [4, "652fef7dd80b934fb2a0e4d6@email.prolific.com"],
    [5, "klrb115@gmail.com"],
]

def get_dialog_for_chat_code(chat_code):
    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()

    # Get all dialogs for the given chat_code
    cursor.execute("""
        SELECT chat_code, utterance, time, user_id
        FROM dialog
        WHERE chat_code = ?
    """, (chat_code,))
    
    dialog_responses = cursor.fetchall()
    conn.close()
    
    if not dialog_responses:
        print(f"No dialog found for chat_code: {chat_code}")
    
    return dialog_responses

def get_chat_codes_for_listener(listener_id):
    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()

    # Get all chat_codes for the listener
    cursor.execute("""
        SELECT chat_code
        FROM session
        WHERE listener_id = ?
    """, (listener_id,))
    chat_codes = [row[0] for row in cursor.fetchall()]
    conn.close()
    return chat_codes

for user in users:
    listener_id = user[0]
    email = user[1]
    print(f"Fetching dialog responses for listener with email: {email}")
    chat_codes = get_chat_codes_for_listener(email)
    all_dialogs = []
    for chat_code in chat_codes:
        dialogs = get_dialog_for_chat_code(chat_code)
        if not dialogs:
            continue
        context = []
        therapist_index = 0
        for i, dialog in enumerate(dialogs):
            user_id = dialog[3]
            if user_id.startswith("_AI_AGENT_"):
                role = "patient"
            else:
                role = "therapist"
            utterance = f"{role}: {dialog[1]}"

            if role == "therapist":
                # Create an entry with the accumulated context and the current therapist response
                dialogue_entry = {
                    "therapist_id": listener_id,
                    "chat_code": dialog[0],
                    "therapist_index": therapist_index,
                    "context": context.copy(),  # All previous utterances
                    "next": [utterance]  # The current therapist response
                }
                all_dialogs.append(dialogue_entry)
                therapist_index += 1

            context.append(utterance)  # Add the current utterance to the context
    
    # Save to JSON file
    if all_dialogs:
        json_filename = f"dialogues_{listener_id}.json"
        with open(json_filename, 'w') as json_file:
            json.dump(all_dialogs, json_file, indent=4)

        print(f"Dialog responses saved to {json_filename}")
    else:
        print(f"No dialogs found for listener with email: {email}")

# This script will save all dialog responses for each chat_code of the specified listeners into JSON files.
```
